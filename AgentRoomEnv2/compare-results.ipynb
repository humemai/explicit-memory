```python
from glob import glob
import pandas as pd
from explicit_memory.utils import read_yaml


results_all = []
for results_path in glob("./training_results/room-size=s/lstm/hp-tuning/*/results.yaml"):
    train_path = results_path.replace("results.yaml", "train.yaml")
    results = read_yaml(results_path)
    train = read_yaml(train_path)
    results_all.append(
        {
            "test_score": results["test_score"]["mean"],
            "val_score": max(results["validation_score"], key=lambda x: x["mean"])["mean"],
            "batch": train["batch_size"],
            "warm": train["warm_start"],
            "replay_buffer": train["replay_buffer_size"],
            "gamma": train["gamma"],
            "hidden_size": train["nn_params"]["hidden_size"],
            "target_update": train["target_update_rate"],
            "ddqn": train["ddqn"],
            "dueling_dqn": train["dueling_dqn"],
            "split_reward": train["split_reward_training"],
            "dir": results_path.split("/")[-2],
        }
    )

df = pd.DataFrame(results_all)
df = df.sort_values("test_score", ascending=False)
df[:30]
```
